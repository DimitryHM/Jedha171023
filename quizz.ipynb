```python
def ask_question(nb_attempts_left, question_sentence, right_answer):
    if nb_attempts_left > 0:
        answer = input(question_sentence)
        answer = answer.lower()
        while answer != right_answer:
            nb_attempts_left -=1
            print("Mauvaise réponse ! Il vous reste {} vies.".format(nb_attempts_left))
            if nb_attempts_left == 0:
                print("GAME OVER")
                break
            answer = input(question_sentence)
            answer = answer.lower()
    print('')
    return nb_attempts_left

questions_answers_list = [
    ("En quelle année est sorti Super Mario Bros. ?", "1985"),
    ("Quel est le nom de la dernière console créée par Sega ?", "dreamcast"),
    ("Quel est le nom de la société responsable, aux débuts, de la série Halo ?", "bungie")
]

nb_attempts = 3

for question, answer in questions_answers_list :
    nb_attempts = ask_question(nb_attempts, question, answer)

if nb_attempts > 0:
    print("Congratulations, you have won the game!")
```
